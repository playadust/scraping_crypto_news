```python
html = browser.page_source;
soup = BeautifulSoup(html, "lxml")

news = []

for div in soup.find_all('div', {'class': 'post boxed'}):
    article = {}
    #print(div)
    article['title'] = div.find('span', {'class': 'postTitle'}).text
    article['date'] = div.find('span', {'class': 'date'}).text
    article['author'] = div.find('span', {'class': 'author'}).text[4:]
    article['lead'] = div.find('p', {'class': 'lead'}).text
    try: article['badge'] = div.find_all('p')[1].text
    except: article['badge'] = ""
    article['url'] = div.find('a').attrs['href']
   
    news.append(article)

df = pd.DataFrame(news)
df
```

```python
df.to_csv('coinTelegraph_feb13.csv')
```
